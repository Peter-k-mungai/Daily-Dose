```python
import zipfile
import requests
from io import BytesIO
import geopandas as gpd

# Step 1: Download and extract the ZIP file
url = "https://jonnyphillips.github.io/Intro_Spatial/Day_2/Data/MCMV_new.zip"
response = requests.get(url)
zip_file = BytesIO(response.content)

# Extract the ZIP file contents to a temporary directory
with zipfile.ZipFile(zip_file, 'r') as z:
    z.extractall("path_to_extracted_folder")

# Step 2: Read the shapefile using geopandas
shapefile_path = "path_to_extracted_folder/MCMV_new.shp"  # Adjust the path if necessary
housing = gpd.read_file(shapefile_path)

# Step 3: Inspect the contents of the shapefile
# Display the first few rows of the GeoDataFrame
print(housing.head())

# Display the column names to understand the structure of the dataset
print("Column names:", housing.columns)

# Display the full summary of the GeoDataFrame
print(housing.info())

# Display the coordinate reference system
print("CRS:", housing.crs)
```

```python
# Filter municipalities in the state of São Paulo (assuming 'state_code' column is correct)
sao_paulo = municipalities[municipalities['UF'] == 'SP']

# Calculate centroids of the municipalities in São Paulo
sao_paulo['centroid'] = sao_paulo.geometry.centroid

# Create a GeoDataFrame for centroids
sao_paulo_centroids = gpd.GeoDataFrame(sao_paulo, geometry='centroid')

# Get the outer boundary of São Paulo
sao_paulo_boundary = sao_paulo.unary_union

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))

# Plot municipalities
sao_paulo.plot(ax=ax, color='white', edgecolor='black')

# Plot centroids
sao_paulo_centroids.plot(ax=ax, color='red', markersize=5, label='Centroids')

# Plot outer boundary
gpd.GeoSeries(sao_paulo_boundary).plot(ax=ax, color='none', edgecolor='blue', linewidth=2, label='Boundary')

# Add title and legend
plt.title("Centroids of Municipalities in São Paulo State")
plt.legend()

# Show the plot
plt.show()
```

```python
# Calculate the mean Human Development Index (HDI) of municipalities in each state
mean_hdi_by_state = municipalities.groupby('UF')['IDHM_10'].mean().reset_index()

# Rename columns for clarity
mean_hdi_by_state.columns = ['State', 'Mean_HDI']

# Display the result
print(mean_hdi_by_state)
```

```python


# Filter for 'Gaucha do Norte' municipality
gaucha_do_norte = municipalities[municipalities['NOME'] == 'Gaucha do Norte']

# Filter for 'Parque do Xingu' indigenous territory
parque_do_xingu = indigenous_groups[indigenous_groups['nome'] == 'Parque do Xingu']

# Perform the intersection to get the area of Gaucha do Norte within Parque do Xingu
intersection = gpd.overlay(gaucha_do_norte, parque_do_xingu, how='intersection')

# Save the intersection as a new shapefile
output_shapefile = r"C:\Users\Admin\Downloads\gaucha_do_norte_in_parque_do_xingu.shp"
intersection.to_file(output_shapefile)

# Display the intersection
print(intersection)
```

```python
acre_housing = housing[housing['UF'] == 'AC']

# Check the filtered dataset
print("Filtered Acre Housing Data:")
print(acre_housing)

# Ensure the geometries are in a projected CRS for distance calculation
acre_housing = acre_housing.to_crs(epsg=32718)  # Use appropriate EPSG code for projected CRS

# Calculate distances between each pair of housing projects
distances = acre_housing.geometry.apply(lambda x: acre_housing.distance(x))

# Find the minimum non-zero distance
min_distance = distances[distances > 0].min().min()

# Find the two closest projects
closest_projects = distances[distances == min_distance].stack().index.tolist()

# Debugging output
print("Distance Matrix:")
print(distances)
print("Minimum Distance:", min_distance)
print("Closest Projects Indices:", closest_projects)

# Ensure that closest_projects contains valid indices
if len(closest_projects) > 0:
    # Extract the indices of the two closest projects
    project_1_idx, project_2_idx = closest_projects[0]

    # Get the two closest projects by their actual indices
    project_1 = acre_housing.iloc[project_1_idx[0]]
    project_2 = acre_housing.iloc[project_1_idx[1]]

    # Create a 10km buffer around each project
    buffer_1 = project_1.geometry.buffer(10000)
    buffer_2 = project_2.geometry.buffer(10000)

    # Create a GeoDataFrame for the buffers
    buffers = gpd.GeoDataFrame({'geometry': [buffer_1, buffer_2]}, crs=acre_housing.crs)

    # Save the buffers to a new shapefile
    output_shapefile = r"C:\Users\Admin\Downloads\acre_housing_buffers.shp"
    buffers.to_file(output_shapefile)

    # Plotting the results
    fig, ax = plt.subplots(figsize=(10, 10))
    acre_housing.plot(ax=ax, color='blue', edgecolor='black', alpha=0.5)
    buffers.plot(ax=ax, color='red', alpha=0.5, edgecolor='black')
    gpd.GeoSeries([project_1.geometry, project_2.geometry]).plot(ax=ax, color='green', markersize=100)
    plt.title("10km Buffers Around Closest Social Housing Projects in Acre")
    plt.show()

    print("Buffers saved to:", output_shapefile)
else:
    print("No valid closest projects found.")
```

```python
# Group by municipality and calculate the total number of housing units (UH)
housing_units_by_municipality = housing.groupby('COD_MUN')['UH'].sum().reset_index()

# Rename columns for clarity
housing_units_by_municipality.columns = ['COD_MUN', 'Total_UH']

# Merge with municipalities shapefile
municipalities = municipalities.merge(housing_units_by_municipality, on='COD_MUN', how='left')

# Fill NaN values with 0 (assuming municipalities with no housing units should be marked as 0)
municipalities['Total_UH'] = municipalities['Total_UH'].fillna(0)

# Identify the municipalities with the lowest and highest number of housing units
lowest_housing_municipality = municipalities.loc[municipalities['Total_UH'].idxmin()]
highest_housing_municipality = municipalities.loc[municipalities['Total_UH'].idxmax()]

# Plotting the distribution of total housing units by municipality
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
municipalities.plot(column='Total_UH', cmap='OrRd', legend=True,
                    legend_kwds={'label': "Total MCMV Housing Units by Municipality",
                                 'orientation': "horizontal"}, ax=ax)

# Highlight the municipalities with the lowest and highest housing units
municipalities[municipalities['COD_MUN'] == lowest_housing_municipality['COD_MUN']].plot(ax=ax, color='blue', markersize=50)
municipalities[municipalities['COD_MUN'] == highest_housing_municipality['COD_MUN']].plot(ax=ax, color='green', markersize=50)

# Add titles and annotations
plt.title("Distribution of Total MCMV Housing Units by Municipality in Brazil")
plt.annotate(f"Lowest: {lowest_housing_municipality['NOME']} ({int(lowest_housing_municipality['Total_UH'])} units)",
             xy=(lowest_housing_municipality.geometry.centroid.x, lowest_housing_municipality.geometry.centroid.y),
             xytext=(10, 10), textcoords='offset points', color='blue')
plt.annotate(f"Highest: {highest_housing_municipality['NOME']} ({int(highest_housing_municipality['Total_UH'])} units)",
             xy=(highest_housing_municipality.geometry.centroid.x, highest_housing_municipality.geometry.centroid.y),
             xytext=(10, 10), textcoords='offset points', color='green')

plt.show()
```
